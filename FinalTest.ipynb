Find the python program to find the odd numbers in the list of numbers entered by user.

```python
numbers = input("Enter a list of numbers (separated by spaces): ")
numbers = [int(x) for x in numbers.split()]

odd_numbers = [x for x in numbers if x % 2 != 0]

print("Odd numbers:", odd_numbers)
```

Write a python & create a class named Employee with attributes name, role, and salary. Write a method to calculate the annual salary of the employee.

```python
class Employee:
    def __init__(self, name, role, salary):
        self.name = name
        self.role = role
        self.salary = salary

    def calculate_annual_salary(self):
        annual_salary = self.salary * 12
        return annual_salary

employee = Employee("Rahul", "Software Engineer", 50000)
print(employee.name)
print(employee.role)
print(employee.salary)
print("Annual salary is",employee.calculate_annual_salary())
```

Given a dataset containing exam scores of students, write a program using NumPy to perform the following statistical analysis: a) Calculate the mean and median of the exam scores. b) Compute the standard deviation and variance of the scores.

```python

import numpy as np

# Load exam scores data
scores = np.array([72,85,80,70,50])

# Calculate mean and median
mean = np.mean(scores)
median = np.median(scores)

# Compute standard deviation and variance
stddev = np.std(scores)
variance = np.var(scores)

print("Mean:", mean)
print("Median:", median)
print("Standard deviation:", stddev)
print("Variance:", variance)

```

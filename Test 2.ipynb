```python
def calculate_bmi(height, weight):
    height_in_meters = height / 100.0

    bmi = weight / (height_in_meters ** 2)
    return bmi

def determine_health_status(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal Weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

height = 175  
weight = 100  
bmi_score = calculate_bmi(height, weight)

health_status = determine_health_status(bmi_score)

print(f"Height: {height} cm")
print(f"Weight: {weight} kg")
print(f"BMI: {bmi_score:.2f}")
print(f"Health Status: {health_status}")
```

```python
import numpy as np

exam_scores = np.array([85, 90, 88, 92, 78, 95, 87, 84, 89, 91])

mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)

print(f"Mean score: {mean_score}")
print(f"Median score: {median_score}")
std_deviation = np.std(exam_scores)
variance = np.var(exam_scores)

print(f"Standard Deviation: {std_deviation}")
print(f"Variance: {variance}")
```
